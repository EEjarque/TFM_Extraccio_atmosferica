# Exploració de les dades auxiliars

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_aux = pd.read_csv("/kaggle/input/abc-database/Level2Data/AuxillaryTable.csv")

# Mostra de les dades
print('Dimensions de les dades: ', df_aux.shape, '\n')
df_aux.head(4)
```

## Estadística bàsica de les característiques planetàries i estel·lars

```python
df_aux.describe()
```

## Distribució de les caracterísitques de les estrelles

```python
# Paràmetres sobre les estrelles
fig, axs = plt.subplots(1, 3, figsize=(14, 3))

#sns.histplot(data=Level2_aux, x="star_distance", kde=True, color="orange", bins=30, log_scale=True, ax=axs[0, 0])
sns.histplot(data=df_aux, x="star_mass_kg", kde=True, color="olive",bins=30, ax=axs[0])
sns.histplot(data=df_aux, x="star_radius_m", kde=True, color="gold",bins=30,log_scale=True, ax=axs[1])
sns.histplot(data=df_aux, x="star_temperature", kde=True, color="teal",bins=30, ax=axs[2])

plt.show()
```

## Distribució de les caracterísitques dels planetes

```python
# Paràmetres sobre els planetes
fig, axs = plt.subplots(2, 3, figsize=(14, 6))

sns.histplot(data=df_aux, x="planet_mass_kg", kde=True, color="orange", bins=30, log_scale=True, ax=axs[0, 0])
sns.histplot(data=df_aux, x="planet_orbital_period", kde=True, color="olive",bins=30,log_scale=True, ax=axs[0, 1])
sns.histplot(data=df_aux, x="planet_distance", kde=True, color="green",bins=30,log_scale=True, ax=axs[0, 2])
sns.histplot(data=df_aux, x="planet_radius_m", kde=True, color="gold",bins=30,log_scale=True, ax=axs[1, 0])
sns.histplot(data=df_aux, x="planet_surface_gravity", kde=True, color="teal",bins=30,log_scale=True, ax=axs[1, 1])
sns.histplot(data=df_aux, x="star_distance", kde=True, color="orange", bins=30, log_scale=True, ax=axs[1, 2])

plt.show()
```

### Característiques dels exoplanetes comparat amb els planetes del Sistema Solar

```python
ax = sns.scatterplot(data=df_aux, x="planet_orbital_period", y="planet_radius_m", alpha=0.1, s=10)
ax.set(xscale="log", yscale="log")
ax.set_xlabel("Període orbital (dies)")
ax.set_ylabel('Radi planetari (m)')

plt.scatter(x=88, y=2.44e6, color='darkorange', alpha = 0.5)
ax.annotate("Mercuri", (88, 2.44e6), color='darkorange')

plt.scatter(x=224, y=6.051e6, color='darkorange', alpha = 0.5)
ax.annotate("Venus", (224, 6.051e6), color='darkorange', ha='left', va='top')

plt.scatter(x=365, y=6370000, color='darkorange', alpha = 0.5)
ax.annotate("La Terra", (365, 6370000), color='darkorange')

plt.scatter(x=686, y=3.389e6, color='darkorange', alpha = 0.5)
ax.annotate("Mart", (686, 3.389e6), color="darkorange")

plt.scatter(x=4332, y=6.99e7, color='darkorange', alpha = 0.5)
ax.annotate("Jupiter", (4332, 6.99e7), color="darkorange")

plt.scatter(x=10747, y=6.03e7, color='darkorange', alpha = 0.5)
ax.annotate("Saturn", (10747, 6.03e7), color="darkorange")

plt.scatter(x=30688, y=25.362e6, color='darkorange', alpha = 0.5)
ax.annotate("Urà", (30688, 25.362e6), color="darkorange", ha='right', va='bottom')

plt.scatter(x=60193, y=24.622e6, color='darkorange', alpha = 0.5)
ax.annotate("Neptú", (60193, 24.622e6), color="darkorange", ha='right', va='top')

plt.show()
```
